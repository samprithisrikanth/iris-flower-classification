```python
data = pd.read_csv("Iris.csv")
```

CONVERTING CSV DATASET IN DATAFRAME

# STEP 4-DATA VISUALIZATION

```python
plt.scatter(df["SepalLengthCm"], df["SepalWidthCm"])

plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")

plt.show()
```

# STEP 5-SEPERATING INPUT AND OUTPUT FEATURES

```python
X = df.drop(["Id","Species"], axis =1) #Input/Train
Y = df["Species"] #Output/Target
```

# STEP 6-SPLIT DATASET INTO TRAINING AND TESTING DATA

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    Y,
    test_size=0.2,
    random_state=42
)
```

# STEP 7-CREATE ML MODEL

```python
model = KNeighborsClassifier(n_neighbors=3)
```

# STEP 8 — TRAIN MODEL

```python
model.fit(X_train, y_train)
```

# STEP 9-MAKE PREDICTIONS

```python
predictions = model.predict(X_test)
```

# STEP 10-CHECK ACCURACY

```python
accuracy = accuracy_score(y_test, predictions)
print("Accuracy:", accuracy)
```

# STEP 11-PREDICT CUSTOM FLOWER

```python
sample = pd.DataFrame(
    [[5.1, 3.5, 1.4, 0.2]],
    columns=[
        "SepalLengthCm",
        "SepalWidthCm",
        "PetalLengthCm",
        "PetalWidthCm"
    ]
)

result = model.predict(sample)
print("Predicted Species:", result[0])
```
